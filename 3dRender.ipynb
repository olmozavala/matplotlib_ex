```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
x = np.linspace(0,np.pi*4)
y = np.sin(x)
```

## 2D to 3D Plot with different z values

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(x, y, 0, zdir='x',color='r', label='x')
ax.plot(x, y, 0, zdir='y',color='g', label='y')
ax.plot(x, y, 0, zdir='z',color='b', label='z')

plt.show()
```
